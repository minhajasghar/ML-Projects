```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import KNNImputer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import joblib
```

```python
train_df = pd.read_csv(r"C:\Users\minha\Downloads\StudentPerformanceFactors.csv")
```

```python
impute = KNNImputer()

for i in train_df.select_dtypes(include="number").columns:
    train_df[i] = impute.fit_transform(train_df[[i]])
```

```python
for i in train_df.select_dtypes(include="object").columns:
    train_df[i] = train_df[i].fillna(train_df[i].mode()[0])
```

```python
train_df.info()
```

```python
train_df.isnull().sum()
```

```python
le = LabelEncoder()

for i in train_df.select_dtypes(include="object").columns:
    train_df[i] = le.fit_transform(train_df[i])
```

```python
train_df = train_df.select_dtypes(include="number").astype("int64")
```

```python
train_df.info()
```

```python
X = train_df.drop(columns=["Exam_Score"])
y = train_df["Exam_Score"]
```

```python
train_x, test_x, train_y, test_y = train_test_split(X,y,test_size=0.2)

print(train_y.shape)  
print(test_y.shape)  
```

```python
model = SVC()
model.fit(train_x,train_y)
```

```python
y_pred = model.predict(test_x)
```

```python
print("Accuracy:",accuracy_score(test_y,y_pred))
```

```python
joblib.dump(model, 'svc_model.pkl')  
print("Model saved successfully!")
```

```python
import joblib
import numpy as np

def preprocess_input(input_data):
    return np.array([
        input_data['Hours_Studied'],
        input_data['Attendance'],
        input_data['Parental_Involvement'],  # 1 for Yes, 0 for No
        input_data['Access_to_Resources'],   # 1 for Yes, 0 for No
        input_data['Extracurricular_Activities'],  # 1 for Yes, 0 for No
        input_data['Sleep_Hours'],
        input_data['Previous_Scores'],
        input_data['Motivation_Level'],
        input_data['Internet_Access'],   # 1 for Yes, 0 for No
        input_data['Tutoring_Sessions'],  
        input_data['Family_Income'],   
        input_data['Teacher_Quality'],   
        input_data['School_Type'],      
        input_data['Peer_Influence'],
        input_data['Physical_Activity'],
        input_data['Learning_Disabilities'],  
        input_data['Parental_Education_Level'], 
        input_data['Distance_from_Home'],
        input_data['Gender']  
    ])

def get_user_input():
    print("Enter the following details:")

    Hours_Studied = float(input("Hours Studied (e.g., 5): "))
    Attendance = float(input("Attendance (%) (e.g., 85): "))

    Parental_Involvement = input("Parental Involvement (Y/N): ").strip().lower()
    Parental_Involvement = 1 if Parental_Involvement == 'y' else 0

    Access_to_Resources = input("Access to Resources (Y/N): ").strip().lower()
    Access_to_Resources = 1 if Access_to_Resources == 'y' else 0

    Extracurricular_Activities = input("Extracurricular Activities (Y/N): ").strip().lower()
    Extracurricular_Activities = 1 if Extracurricular_Activities == 'y' else 0

    Sleep_Hours = float(input("Sleep Hours (e.g., 6): "))
    Previous_Scores = float(input("Previous Scores (e.g., 80): "))
    Motivation_Level = float(input("Motivation Level (e.g., 7): "))

    Internet_Access = input("Internet Access (Y/N): ").strip().lower()
    Internet_Access = 1 if Internet_Access == 'y' else 0

    Tutoring_Sessions = input("Tutoring Sessions (Y/N): ").strip().lower()
    Tutoring_Sessions = 1 if Tutoring_Sessions == 'y' else 0

    Family_Income = input("Family Income (High/Low): ").strip().lower()
    Family_Income = 1 if Family_Income == 'high' else 0

    Teacher_Quality = input("Teacher Quality (Good/Bad): ").strip().lower()
    Teacher_Quality = 1 if Teacher_Quality == 'good' else 0

    School_Type = input("School Type (Public/Private): ").strip().lower()
    School_Type = 1 if School_Type == 'private' else 0

    Peer_Influence = float(input("Peer Influence (e.g., 7): "))
    Physical_Activity = float(input("Physical Activity (e.g., 4): "))

    Learning_Disabilities = input("Learning Disabilities (Y/N): ").strip().lower()
    Learning_Disabilities = 1 if Learning_Disabilities == 'y' else 0

    Parental_Education_Level = input("Parental Education (Y/N - Did your parent complete higher education?): ").strip().lower()
    Parental_Education_Level = 1 if Parental_Education_Level == 'y' else 0

    Distance_from_Home = input("Do you live far from school? (Y/N): ").strip().lower()
    Distance_from_Home = 1 if Distance_from_Home == 'y' else 0

    Gender = input("Gender (Male/Female): ").strip().lower()
    Gender = 1 if Gender == 'female' else 0

    input_data = {
        'Hours_Studied': Hours_Studied,
        'Attendance': Attendance,
        'Parental_Involvement': Parental_Involvement,
        'Access_to_Resources': Access_to_Resources,
        'Extracurricular_Activities': Extracurricular_Activities,
        'Sleep_Hours': Sleep_Hours,
        'Previous_Scores': Previous_Scores,
        'Motivation_Level': Motivation_Level,
        'Internet_Access': Internet_Access,
        'Tutoring_Sessions': Tutoring_Sessions,
        'Family_Income': Family_Income,
        'Teacher_Quality': Teacher_Quality,
        'School_Type': School_Type,
        'Peer_Influence': Peer_Influence,
        'Physical_Activity': Physical_Activity,
        'Learning_Disabilities': Learning_Disabilities,
        'Parental_Education_Level': Parental_Education_Level,
        'Distance_from_Home': Distance_from_Home,
        'Gender': Gender
    }
    
    return input_data

def main():
    model = joblib.load('svc_model.pkl')

    user_input = get_user_input()
    processed_input = preprocess_input(user_input)

    prediction = model.predict([processed_input])
    print(f"Predicted performance measure: {prediction[0]}")

if __name__ == "__main__":
    main()
```
